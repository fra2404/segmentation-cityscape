```python
# Import the download script (assuming it's in the same directory or uploaded)
# If cloned from git, adjust path if needed
from download import login, get_available_packages, download_packages
```

```python
# Login to Cityscapes
session = login()
print("Logged in successfully.")
```

```python
# List available packages
packages = get_available_packages(session=session)
print("Available packages:")
for p in packages:
    info = f" {p['name']} -> {p['size']}"
    if p.get('thirdparty') == '1':
        info += " (third party)"
    print(info)
```

```python
# Set destination path
try:
    import google.colab
    dest_path = "/content/drive/MyDrive/cityscapes"
except ImportError:
    dest_path = "./data/cityscapes"  # Local path

os.makedirs(dest_path, exist_ok=True)
print(f"Destination: {dest_path}")
```

```python
# Download the main packages (adjust names based on the list above)
# Typically: gtFine_trainvaltest.zip and leftImg8bit_trainvaltest.zip
to_download = [
    "gtFine_trainvaltest.zip",
    "leftImg8bit_trainvaltest.zip"
]

download_packages(session=session,
                  package_names=to_download,
                  destination_path=dest_path,
                  resume=True)
print("Download completed!")
```
